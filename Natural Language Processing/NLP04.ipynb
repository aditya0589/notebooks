# **NLP04 Text Classification with Machine Learning**

## **Text Classification**
Text classification is a supervised learning problem where the given text is to be assigned to one of the many predefined classes.

### Types of Text Classification:
There are three main types of text classification:
1. Binary classification: email spam classifier
2. Multiclass classifier : genres for news articles
3. Multilabel classification: multiple label classification.

Text classification is the most widely used Natural Language processing task.


### **Applications using Text classification:**
1. Email spam classification system
2. customer support systems
3. Sentiment Analysis
4. Language detection systems
5. Misinformation and fake news detection


## **Classification Pipeline:**
1. Data aquisition
2. Text preprocessing
3. Text Vectorization
4. Modelling and Evaluation
5. Deployment and Montoring



Now lets perform the text classification on the highly popular IMDB movie reviews dataset using machine learning

```python
import numpy as np
import pandas as pd
```

```python
temp_df = pd.read_csv('IMDB Dataset.csv')
```

```python
df = temp_df.iloc[:10000]
```

we only use the first 10,000 rows for the model. (inorder to increase the speed of the process. However in real world scenarios, you would want to use the entire data)

We would then go ahead and do some basic preprocessing

```python
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)  #drops all the duplicate rows in the data
```

The following basic preprocessing steps include:
1. Removing HTML tags
2. lowercasing
3. removing stop words

```python
import re
def remove_tags(raw_text):
    cleaned_text = re.sub(re.compile('<.*?>'), '', raw_text)
    return cleaned_text
```

```python
df['review'] = df['review'].apply(remove_tags)
```

```python
df
```

```python
df['review'] = df['review'].apply(lambda x:x.lower())
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

sw_list = stopwords.words('english')

df['review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

```python
X = df.iloc[:,0:1]
y = df['sentiment']
```

```python
X
```

```python
y
```

### Modelling
We use the Guassian Naive Baye's model and the Random Forest Model for the classification

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

Bag of words vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_train_bow.shape
```

Guassian Naive Baye's

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

gnb.fit(X_train_bow,y_train)
```

```python
y_pred = gnb.predict(X_test_bow)

from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

```python
cv = CountVectorizer(max_features=3000)

X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

Similarly, the above process can be done by using Tf-Idf or the Word2Vec vectorization models
