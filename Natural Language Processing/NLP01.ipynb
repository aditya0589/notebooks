## **Removing HTML Tags**

When you want to process data taken from web pages, you often find them being accompanied by certain HTML tags. These tags are important for the browser to render the web page but are needed to be removed for language processing tasks.

```python
text = df['review'][0].lower()
text
```

In the above text, you find various HTML tags like < br >< /br > tags. These should be removed.

You can do it by using **Regular Expressions**.

```python
import re

def remove_html_tags(text: str) -> str:
    clean_text = re.sub(r'<.*?>', '', text)
    return clean_text

removed_text = remove_html_tags(text)

print(removed_text)
```

The above python function uses regular expressions to remove any HTML tags in a given string

```python
df['review'] = df['review'].apply(remove_html_tags) # removes HTML tags for full dataframe
```

## **Removing URLs**
You can remove URLs by using **Regular Expressions**

Consider the below text:

```python
text = """Last week I found a really useful collection of open datasets that helped me start a new machine learning project. The resources are neatly organized by domainhealthcare, finance, education, and many moreand theyre completely free to use. If youre interested, you can explore them at https://example.com/datasets
 and download whatever you need to begin experimenting."""
```

```python
import re

def remove_urls(text: str) -> str:
    # Regex to detect URLs
    url_pattern = r'(https?://\S+|www\.\S+|\S+\.\w{2,}(\/\S*)?)'
    clean_text = re.sub(url_pattern, '', text)
    return clean_text.strip()
```

```python
text = remove_urls(text)
print(text)
```

```python
import re
```

```python
def expand_abbreviations(text: str) -> str:
    # Regex word boundary replacement
    def replace(match):
        word = match.group(0).lower()
        return genz_abbreviations.get(word, word)

    pattern = r'\b(' + '|'.join(map(re.escape, genz_abbreviations.keys())) + r')\b'
    return re.sub(pattern, replace, text, flags=re.IGNORECASE)
```

```python
sentence = "idk lol, ngl this song is fire btw"
result = expand_abbreviations(sentence)
print(result)
```

## **Removing Stop words**

Stop words like a, an, the do not always contribute to the meaning of a sentence. SO having those words make no sense for certain situations. They might have to be removed.


```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words('english')
```

The above are the various stop words of the english language

```python
stopwords.words('spanish')
```

The above are the stopwords of spanish language

```python
def remove_stopwords(text):
    words = text.lower().split()
    cleaned = [w for w in words if w not in stopwords]
    return " ".join(cleaned)
```

```python
example = "The United States is an enemy of the russian federation"
```

```python
final = remove_stopwords(example)
final
```

## **Handling emojis**

People often tend to use emojis to express their emotions. However, most learning algoithms do not understand emojis. Therefore it is import to handle them.

Emojis can be handled in two ways:
1. Remove emojis from the data
2. Replace the emoji with its textual meaning

**Removing Emojis**

```python
import re

def remove_emojis(text):
    # Regex pattern for emojis (covers most unicode emoji ranges)
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags
        "\U00002702-\U000027B0"  # dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r"", text)
```

```python
sentence = "I loved this movie. It was "
```

```python
sentence
```

```python
sentence = remove_emojis(sentence)
sentence
```

## **Stemming**

**What is inflection in Language:**

in grammer, inflection is the modification of a word to express various grammatical categories such as tense, case, voice, aspect, person, number, gender or mood

ex: eat, ate, eating represent the same action but at different periods of time.
<br>
good, better, best represent the same thing but at different levels of intensity.

**Stemming**

Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the language.

ex: eat, ate, eating are all mapped to the word eat.

Stemming is most useful in Information retrieval systems

**Stemming using NLTK stemmers**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()
```

```python
words = ["running", "jumps", "easily", "fairly", "studies", "studying"]

for w in words:
    print(w, " --> ", stemmer.stem(w))
```

```python
word_list = ["connection", "connected", "connecting", "connections"]

for word in word_list:
    print(f"{word:12} -> {stemmer.stem(word)}")

```

You may find that sometimes it may not be very accurate, and you might need to manually fix these

In stemming, the obtained word might not always be a word in the language as you have seen in example 1. If you strictly want the derived word to be part of the language, you can use another hyet similar method called as Lemmatization

**What is Lemmatization:**

Lemmatization, unlike stemming, reduces the inflicted word properly ensuring that the root word belongs to the language. In lemmatization, the root word is called the ***Lemma***.

A lemma (prural for lemmas or lemmata) is the canonical form, dictionary form or a citation form of a set of words.

However this makes the process of lemmatization much slower than stemming

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
```

```python
words = ["running", "jumps", "better", "studies", "leaves"]

for w in words:
    print(w, " --> ", lemmatizer.lemmatize(w))
```

```python
print(lemmatizer.lemmatize("running", pos="v"))
print(lemmatizer.lemmatize("better", pos="a"))
```

In NLTK lemmatization, the pos parameter stands for Part of Speech.
It tells the lemmatizer what type of word (noun, verb, adjective, adverb, etc.) it is dealing with.

**Why is pos important?**

Many words change meaning depending on their grammatical category.
Without pos, the lemmatizer assumes noun (pos='n'), which can give incorrect or unchanged results.

| POS Code | Meaning   | Example           |
| -------- | --------- | ----------------- |
| `"n"`    | Noun      | car, apple        |
| `"v"`    | Verb      | run, jump         |
| `"a"`    | Adjective | good, better      |
| `"r"`    | Adverb    | quickly, silently |


```python
sentence = "The children are running and playing happily in the parks."
tokens = word_tokenize(sentence)

lemmas = [lemmatizer.lemmatize(word, pos="v") for word in tokens]
print(lemmas)
```
