A production machine Learning system is more than just the Machine Learning model. It is the combination of the model, the data, the interface, the infastructure and the underlying hardware.

**First Question to ask while building an ML system**: When and when not to use Machine Learning.

## **When to use Machine Learning**:

ML can be used if your system has the following characteristics:

1. **Learn**: The system has the ability to learn
    ex: RSBMS is not a machine learning system
    The system would require data to learn from

2. **Complex**: Pattens are complex. For simple patterns which can be solved by just a few if-else statements, machine learning is an overkill.
  ex: To predict the zip-code of a region, we dont need to use ML. We can simply look up from a database.

3. **Patterns**: There must be patterns to learn.
  ex: Dont make an ML system to predict the next outcome of a fair die.

4. **Existing data**: Data must be available or should be possible to collect.

5. **predictions**: It is a predictive problem.
    ex: Instead of finding exact solutions to computationally intensive problems, we can just use Machine Learning to predict the approximate answer by much less cost.

6. **Unseen data**: The unseen data in production must share the same patterns with the training data. (ie: both the data must have the same distribution)

Machine Learning works well if the problem has the following additional characteristics.

1. Repititive task
2. The cost of wrong predictions is cheap
3. The predictions are at scale.
4. If the patterns are constantly changing. (static patterns can be modelling using normal programming, no ML needed)

## **Machine Learning in Research vs in Production**

Machine Learning in research is very different from Machine Learning in production.

| Category | Research | Production |
|----------|----------|------------|
| Requirements | State-of-the-art model performance on benchmark datasets | Different stakeholders have different requirements |
| Computational Priority | Fast training, high throughput | Fast inference, low latency |
| Data | Static | Constantly shifting |
| Fairness | Good to have (sadly) | Important |
| Interpretability | Good to have | Important |


<br>

### Different Stakeholders have different Requirements:
1. **ML Engineers** want the model to recommend the restaurants that most users would likely want to go to.

2. **Sales teams** would want the model to recommend the most expensive restaurant

3. **Product team** would want a model which has the least possible latency

4. **ML Platform teams** would prioritize upgrading the system over the model.


<br>

### Computational priorities:
In research people focuss more on training better models, so they must have faster training.

As production deals with end users, they must need faster inference.

<br>

### Latency vs Throughput
Research prioritizes higher throughput ie: more requestsmust be processed in less time. wheras production focusses on low latency.

On the other hand, sometimes high latecy and slower requests might be important in production.

ex: In Amazon, slower requests mean more data in the account because they made many purchases. ie: A valuable customer.

<br>

### Data
During research, we use static datasets which are often clean and well formatted.

In production, data is messy, noisy, probably unstructured and constantly shifting.

It might be biased and we might not know how its biased.

If you work with user data, you will have to worry about privacy and regulatory concerns.

image.png

<br>

### Fairness
It doesnt exist in Research but very important in production.

ex: An ML system might reject a loan to a worthy candidate just because he is black. As the model training data had a lot of black defaulters.

<br>

### Interpretability
In 2020, nobel Laurete Geoffery Hinton proposed a heated debate about the importence of interpretability in ML systems.


“Suppose you have cancer and you have to choose between a black box
AI surgeon that cannot explain how it works but has a 90% cure rate and
a human surgeon with an 80% cure rate. Do you want the AI surgeon to
be illegal?”

Most peple actually voted to select the human surgeon.

interpretability is important for users, both business leaders and endusers, to understand why a decision is made so that they can trust a model



## **Machine Learning vs Traditional Software**

              "Machine Learning is Software Engineering"

However, many challenges are unique to ML alone:
1. In SWE, we focuss on modularity and seperation of concerns. On the contrary, ML systems are part code, part data, part artifacts. hence seperation of concerns fails.

2. In SWE, we only version and test code. But in ML, we need to version and test the data too.

**Unique production challenges appear in ML alone**:

1. Increasing compute over the years.
2. Getting large models to run in production on edge devices is a massive engineering challenge.
3. Models run fast enough to be useful.

  ex: A perfectly accurate auto-completion model is useless if prediction time is greater than the time needed to type.

## **Designing Machine Learning systems in Production**

### **Requirements for ML systems**:

1. **Reliability**

Systems should continue to perform the correct function at the desired level of performance even in the face of adversity.

ex: Software systems crash with exit codes, but ML systems just give wrong predictions. This silent failure of ML systems is difficult to verify and debug.

2. **Scalability**

Your system might work with a simple logistic regression model that fits into an AWS free tier with 1 GB RAM. but as time passes, a complex neural network with 16 GB RAM might be needed.

Your Ml system derving 10,000 users might need to support a million users as the company grows.

3. **maintanability**

Many people work in an ML system:
1. ML Engineers
2. Data Scientists
3. DevOps teams
4. Subject matter experts


Each team should be able to work on the system with their preferred tools and the system must support this. Every stakeholder must feel confortable working with the system.

4. **Adaptability**

The system should adopt to the change in data distribution and business requirements.

5. **Iterative process**:

As Ml systems fails or underperforms, the models must be retrained with the new conditions.
