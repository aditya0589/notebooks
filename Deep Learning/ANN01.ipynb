# **DL02 Artificial Neural Networks**
An Artificial Neural Network (ANN) is a system made up of many simple decision units (neurons) connected together, inspired by how the human brain works, to learn patterns from data and make predictions.

Think of an ANN as a team of perceptrons working together.
1. Each perceptron makes a small decision
2. These decisions are passed through multiple layers

Together, they solve complex problems that a single perceptron cannot

Why do we need ANN instead of a perceptron?

**A single perceptron:**

1. Can only draw a straight line to separate data

2. Works only for simple, linear problems

3. Fails on problems like XOR, image recognition, speech, etc.

**An Artificial Neural Network:**

1. Has hidden layers between input and output

2. Uses non-linear activation functions

3. Can learn curves, shapes, patterns, and complex relationships

## **Architecture of an ANN**

MLP structure:
- **Input Layer**: Receives raw features (e.g., pixel values, sensor data)
- **Hidden Layers**: Intermediate layers that transform features into higher-level representations
- **Output Layer**: Produces final prediction or class scores  
Each neuron in a layer is connected to every neuron in the next layer (*fully connected*). :contentReference[oaicite:3]{index=3}

**How information flows**
Information flows in one direction: **Input  Hidden Layers  Output.**  
This is called a **feed-forward network**, meaning there are no loops or feedback connections in the basic form of ANN

At each neuron:

1. Inputs are multiplied by weights

2. Summed with a bias

3. Passed through a non-linear activation function

This helps the network represent complex relationships.




## **Key concepts of an ANN**

1. **Weights and biases**

Neurons learn weights and biases during training.  
- Weights control the impact of each input signal  
- Bias allows shifting the activation function to better fit data  
These are learned via optimization (e.g., gradient descent).

2. **Activation functions**
Non-linear functions (like sigmoid, ReLU) are applied after summation so the network can model nonlinearities. Without activation functions, the network collapses into a linear model.



## **Forward Propogation in ANN**

**Forward propagation** is the process by which input data moves **forward through a neural network**, layer by layer, to produce an output. At each layer, the network combines inputs with learned weights, adds a bias, applies an activation function, and passes the result to the next layer.

---

### Step-by-Step Mathematical Derivation

Consider a neural network with:
- Input vector: $\mathbf{x} = [x_1, x_2, \dots, x_n]$
- Weights: $\mathbf{W}$
- Bias: $\mathbf{b}$
- Activation function: $f(\cdot)$

---

### 1. Linear Combination (Weighted Sum)

For a neuron, the input is combined as:

$$
z = \sum_{i=1}^{n} w_i x_i + b
$$

In vector form:

$$
z = \mathbf{W} \cdot \mathbf{x} + b
$$

---

### 2. Activation Function

The weighted sum is passed through a non-linear activation function:

$$
a = f(z)
$$

This non-linearity allows the network to learn complex patterns.

---

### 3. Forward Propagation Through Layers

For layer $l$:

$$
z^{(l)} = W^{(l)} a^{(l-1)} + b^{(l)}
$$

$$
a^{(l)} = f\left(z^{(l)}\right)
$$

Where:
- $a^{(0)} = x$ (input layer)
- $a^{(l)}$ is the output of layer $l$

---

### 4. Output Layer

After passing through all layers, the final output is:

$$
\hat{y} = a^{(L)}
$$

This $\hat{y}$ is the networks prediction.

---

### Key Intuition

- Forward propagation = **compute prediction**
- It uses current weights and biases
- No learning happens here (learning happens during backpropagation)
