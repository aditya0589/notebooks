## **DL03 Customer Churn Prediction Example using ANN**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential  # Used to build the ANN
from tensorflow.keras.layers import Dense  # Used to create the hidden layers
```

```python
model = Sequential()
```

```python
# build the layers of the neural network
model.add(Dense(5, activation='relu', input_dim=11)) # hidden layer 1
model.add(Dense(5, activation='relu'))  # hidden layer 2
model.add(Dense(1, activation='sigmoid'))  # The output layer
```

```python
model.summary()   # view the summary of the network
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train,
    y_train,
    batch_size=50,
    epochs=100,
    verbose=1,
    validation_split=0.2
)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
